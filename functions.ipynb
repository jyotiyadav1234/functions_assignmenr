1. What is the difference between a function and a method in Python?

```python

''''Function: A function is a block of code that performs a specific task and can be called independently.
Example:'''
def add(a, b):
    return a + b
print(add(2, 3))

'''Method: A method is a function that is associated with an object and is called on that object.
Example:'''
class Calculator:
    def add(self, a, b):
        return a + b
calc = Calculator()
print(calc.add(2, 3))
```

6. Explain the concept of generators in Python and how they are defined.

```python
'''Generators are functions that yield values one at a time using the yield keyword. Example:'''

def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

for number in count_up_to(5):
    print(number)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

```python
'''map(): Applies a function to each item in an iterable. Example:'''

numbers = [1, 2, 3]
doubled = map(lambda x: x * 2, numbers)
print(list(doubled))

'''reduce(): Applies a function cumulatively to items in an iterable, reducing it to a single value (from functools). Example:'''

from functools import reduce

numbers = [1, 2, 3, 4]
sum_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_numbers)

'''filter(): Filters items in an iterable based on a function that returns True or False. Example:'''

numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(numbers):
  if numbers <2:
    return False
  for i in range(2, int(numbers ** 0.5)+1):
    if numbers % i ==0:
      return False
    return True

for num in range(1,201):
  if is_prime(num):
    print(f"{num} is a prime number.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibo:
    def __init__(self,n):
        self.n =n
        self.count = 0
        self.a , self.b = 0,1

    def __iter__(self):
        return self

    def __next__(self):
      if self.count>= self.n:
        raise StopIteration
      self.a, self.b = self.b , self.a +self.b
      self.count +=1
      return self.a


fib_sequence = Fibo(10)  # Generate the first 10 Fibonacci numbers
for num in fib_sequence:
    print(num)

```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def power(n):
  for i in range(n):
    yield 2** i

for num in power(5):
  print(num)

```

11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:







Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
order is smaller than 100,00 .

Write a Python program using lambda and map.

```python
orders = [
    [1, 10, 5.0],
    [2, 3, 35.0],
    [3, 7, 12.0]
]

adjusted_totals = list(map(lambda order: (
    order[0],
    order[1] * order[2] + (10 if order[1] * order[2] < 100 else 0)
), orders))

print(adjusted_totals)

```
